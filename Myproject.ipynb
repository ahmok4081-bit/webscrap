```python
#creat the loop by useing the function to reuseble
import random

def pig_game():
  max_value = 6
  min_value = 1 # Changed from 0 to 1 for standard dice
  dice_value = random.randint(min_value, max_value)
  return dice_value

dice_value = pig_game() # Call the function and assign its result outside the function
print(f"the dice value is {dice_value}")

#check the number of the player and prevent from the error when they type srtring
while True:
  try:
    user_input_str = input('enter the number of the player (2-4): ')
    user_input = int(user_input_str)
    if 2 <= user_input <= 4:
      break
    else:
      print('Invalid input: please enter a number from 2 to 4.')
  except ValueError:
    print('Invalid input: please enter an integer.')
#creat a list to store the number of the user and return it to 0
player_score= [0 for _ in range(user_input )]
print(player_score)

##condition of the game
#player turn
for player_index in range(user_input):
  print(f'\n--- Player {player_index + 1} turn ---')
  current_score_for_turn = 0 # Initialize score for the current player's turn
  # Loop for the current player's rolling turn
  while True:
    shoud_continue = input('roll the dice? (y/n): ').lower()
    if shoud_continue == 'n':
      player_score[player_index] += current_score_for_turn # Add turn score to total score
      print(f"Player {player_index + 1} holds. Current total score: {player_score[player_index]}")
      break # End player's turn
    elif shoud_continue == 'y':
      dice_value = pig_game() # Roll the dice
      print(f"You rolled a {dice_value}")

      if dice_value == 1:
        print('You rolled a 1! You lost all score for this turn.')
        current_score_for_turn = 0 # Reset turn score
        break # End player's turn
      else:
        current_score_for_turn += dice_value
        print(f"Score for this turn: {current_score_for_turn}")
        print(f"Current total score: {player_score[player_index] + current_score_for_turn}")
    else:
      print('Invalid input. Please enter \'y\' or \'n\'.')

# After all players have had their turn (this simple loop assumes one round, more rounds would be needed for a full game)
# You would then add logic to check for a winner or continue rounds.
```

```python
import random
max_value = 12
min_value = 3
oparator = ['+','-','*']

def generate_theproblem():
    left = random.randint(min_value, max_value)
    right = random.randint(min_value, max_value)
    oparators = random.choice(oparator)
    expr = str(left)+ oparators +str(right)
    answer = eval(expr)
    return expr,answer
expr,answer = generate_theproblem()
print(expr,answer)
```
